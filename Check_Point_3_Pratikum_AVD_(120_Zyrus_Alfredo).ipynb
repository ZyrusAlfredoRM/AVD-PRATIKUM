### **Mengubah Tipe Data**
Karena sebelumnya Object diubah menjadi Float

```python
columns_to_convert = ['Units Sold', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']

for col in columns_to_convert:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('[\$,]', '', regex=True)
    df[col] = df[col].str.replace(r'\((.*?)\)', r'-\1', regex=True)
    df[col] = df[col].str.replace('-', '0')
    df[col] = df[col].astype(float)

print(df.dtypes)
```

```python
df.dtypes
```

```python
results = []

cols = df.select_dtypes(include=['float64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

### **Outliers**

```python
columns_to_impute = ["Gross Sales", "Discounts", "Sales", "COGS", "Profit"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Construct Data**

Melakukan Construct Data (Penambahan Kolom) berdasarkan studi kasus yang saya analisis sebagai berikut:
1. Mencari tau seberapa besar keuntungan dari penjualan
2. Seberapa besar diskon yang diberikan dibandingkan total penjualan
3. Berapa besar biaya produksi terhadap total penjualan

```python
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
df['Discount Rate (%)'] = (df['Discounts'] / df['Gross Sales']) * 100
df['COGS Percentage (%)'] = (df['COGS'] / df['Sales']) * 100

print(df.head())
```

```python
df
```

```python
def categorize_discount(rate):
    if rate < 5:
        return 'Low Discount'
    elif 5 <= rate <= 15:
        return 'Medium Discount'
    else:
        return 'High Discount'

df['Discount Category'] = df['Discount Rate (%)'].apply(categorize_discount)
```

# **Matplotlib & Seaborn**

**Menampilkan segmen mana yang profit nya paling untung/besar - Bar Chart**

Dapat dilihat pada diagram batang dibawah ini, menunjukkan bahwa segmen "Government" memiliki profit yang tertinggi dibanding segmen yang lainnya.

```python
plt.figure(figsize=(8, 5))
df.groupby('Segment')['Profit'].sum().plot(kind='bar', color='red')
plt.title('Total Profit per Segmen Pasar')
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```

**Menampilkan negara mana yang memiliki total penjualan tertingi - Bar Chart**

Pada diagram batang di bawah, menunjukkan bahwa negara yang memiliki total penjualan tertinggi ialah USA, Canada, dan France. Sedangkan Germany dan mexico relatif di bawah.

```python
plt.figure(figsize=(8, 5))
df.groupby('Country')['Sales'].sum().plot(kind='bar', color='green')
plt.title('Total Sales per Negara')
plt.xlabel('country')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

**Menampilkan Profit Berdasarkan Rentang Diskon - Line Chart**

Pada line chart di bawah menunjukkan tren profit berdasarkan diskon. Terlihat bahwa, profit meningkat seiring bertambahnya diskon, meskipun ada penurunan akan tetapi melonjak kembali. Penurunan tersebut bisa terjadi karena efek Over Discounting (Diskon Berlebihan) sehingga menyebabkan Profit menurun.

```python
df['Discount_Range'] = pd.cut(df['Discounts'], bins=10)
discount_profit = df.groupby('Discount_Range')['Profit'].mean().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(discount_profit['Discount_Range'].astype(str), discount_profit['Profit'], marker='o', linestyle='-', color='b')
plt.xticks(rotation=45)
plt.xlabel("Rentang Diskon")
plt.ylabel("Rata-rata Profit")
plt.title("Tren Profit berdasarkan Rentang Diskon")
plt.grid(True)
plt.show()
```

**Menampilkan Korelasi Antar Variable (Numerik) - Heatmap**

Dari heatmap di bawah ini dapat disimpulkan bahwa.
1. Diskon dapat meningkatkan Profit serta Penjualan secara keseluruhan.
2. Semakin tinggi total penjualan maka semakin besar biaya produksi.


```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Antar Variabel")
plt.show()
```

**Kesimpulan:**
Dari visualisasi Bar Chart, Line Chart, dan Heatmap yang sudah tertera di atas menunjukkan bahwa terdapat hubungan yang kuat antara diskon dan profit. Namun pada moment tertentu terdapat penurunan profit meskipun diskon tergolong tinggi, dan hal itu bisa disebabkan oleh banyak faktor termasuk biaya produksi. Terlihat juga harga jual, total penjualan, dan biaya produksi memiliki korelasi/hubungan yang kuat terhadap profit (bukan hanya diskon).
